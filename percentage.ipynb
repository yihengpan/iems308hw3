```python
import re
import nltk
from nltk.corpus import stopwords
stop = stopwords.words('english')
from nltk.corpus import wordnet
import io
import os
import glob
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
```

```python
path1 = '2013'
path2 = '2014'
```

```python
def read_all_files(path):
    criteria = os.listdir(path)
    if len (criteria) == 0:
        raise Exception ('This is an empty file')
    else:
        files = glob.glob(path + '/*')
        frame = []
        for f in files:
            text = open(f, 'r', encoding = 'latin-1').read()
            frame.append(text)
            return frame
```

```python
#sentence tokenize
c2013 = read_all_files(path1)
sentence13 = []
for item in c2013:
    sentence13.extend(sent_tokenize(item))
```

```python
#word tokenize
Tokens = []
for sent in sentence13:
    Tokens.append(nltk.word_tokenize(sent))
Words_List = [nltk.pos_tag(Token) for Token in Tokens]
```

```python
#find all the nouns
stop1 = [',',';','.','%']
Nouns_List = []
for List in Words_List:
    for Word in List:
        if re.match('[NN.*]',Word[1]):
            if Word[0] not in stop:
                if Word[0] not in stop1:
                    Nouns_List.append(Word[0])
```

```python
percentage = []
number = []
for List in Words_List:
    for Word in List:
        if re.match('[CD]',Word[1]):
            index = List.index(Word)
            if re.match('[%]',List[index+1][0]):
                number = Word[0] + '%'
                percentage.append(number)
```

```python
pencentage1 = []
number1 = []
words_list = []
stop0 = ['That','An','This','The','&','And']
for List in Words_List:
    for Word in List:
        if re.match('[CD]',Word[1]) and Word[0] not in stop and Word[0] not in stop0:
            index = List.index(Word)
            if re.match('[percentage]',List[index+1][0]):
                number1 = Word[0] + ' ' + 'percentage'
            pencentage1.append(number1)
```

```python
file = open('percentage.txt','w')
for item in percentage:
    item = item.encode('utf-8')
    file.write("%s"" " % item)
```
